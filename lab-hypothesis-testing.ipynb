In this challenge, we will be working with pokemon data. The data can be found here:

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
df_dragon = df[df["Type 1"]=="Dragon"]["HP"]
df_grass = df[df["Type 1"]=="Grass"]["HP"]
```

```python
#Set the hypothesis

#H0: mu_hp dragon > mu_hp grass
#H1: mu_hp dragon < mu_hp grass

#significance level = 0.05

st.ttest_ind(df_dragon,df_grass, alternative='less')
```

**Findings** : since the pvalue is higher than the significant level, we accept the hypothesis that Pokemons of type Dragon have on average more HP stats than Grass.

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
import pandas as pd
from scipy.stats import f_oneway

#Set the hypothesis

#H0: legendary_pokemon_stats = non_legendary_pokemon_stats
#H0: legendary_pokemon_stats != non_legendary_pokemon_stats

#Lets choose significance level of 5%
alpha = 0.05

# List of stats columns
stats_columns = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

# Perform ANOVA test for each stat
for stat in stats_columns:
    legendary_pokemon = df[df['Legendary'] == True][stat]
    non_legendary_pokemon = df[df['Legendary'] == False][stat]
    
    # Perform one-way ANOVA test
    f_statistic, p_value = f_oneway(legendary_pokemon, non_legendary_pokemon)
    
    # Print results
    print(f'Stat: {stat}')
    print(f'F-statistic: {f_statistic:.2f}')
    print(f'p-value: {p_value:.4f}')
    if p_value < 0.05:
        print('Result: Reject null hypothesis - Significant difference found')
    else:
        print('Result: Fail to reject null hypothesis - No significant difference found')
    print('-' * 40)
```

Based on the p-values obtained, which is less than 0.05, we reject the null hypothesis and conclude that there is a statistically significant difference in the mean stats between Legendary and Non-Legendary PokÃ©mon.
